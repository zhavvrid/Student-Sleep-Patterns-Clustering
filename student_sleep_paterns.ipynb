```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler, normalize
from sklearn.decomposition import PCA

from sklearn.cluster import KMeans
from kneed import KneeLocator
```

```python
df = pd.read_csv('student_sleep_patterns.csv')
```

```python
df
```

```python
df = df.drop(['Student_ID', 'Weekday_Sleep_Start', 'Weekend_Sleep_Start', 'Weekday_Sleep_End', 'Weekend_Sleep_End'],
             axis=1)
```

```python
df['University_Year'] = df['University_Year'].str.replace('st Year', '')
df['University_Year'] = df['University_Year'].str.replace('nd Year', '')
df['University_Year'] = df['University_Year'].str.replace('rd Year', '')
df['University_Year'] = df['University_Year'].str.replace('th Year', '')
```

```python
df.dtypes
```

```python
df['University_Year'] = df['University_Year'].astype(int)
```

```python
df
```

```python
X = pd.get_dummies(df)
```

```python
scaler = StandardScaler()

scaled_X = scaler.fit_transform(X)
```

```python
normalized_X = normalize(scaled_X) 

normalized_X = pd.DataFrame(normalized_X) 
```

```python
pca = PCA(n_components = 2) 

X_pca = pca.fit_transform(normalized_X) 
X_pca = pd.DataFrame(X_pca) 

X_pca.columns = ['P1', 'P2'] 

X_pca
```

```python
plt.figure(figsize=(8,6), dpi=150)
plt.scatter(X_pca['P1'], X_pca['P2'])
plt.show()
```

```python
inertia = []
k_range = range(1, 11)
```

```python
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_pca)
    inertia.append(kmeans.inertia_)
```

```python
knee_locator = KneeLocator(k_range, inertia, curve="convex", direction="decreasing")
optimal_k = knee_locator.elbow

print(f"Оптимальное количество кластеров: {optimal_k}")
```

```python
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o', linestyle='--')
plt.xlabel("Количество кластеров")
plt.ylabel("Инерция")
plt.title("Метод локтя для выбора количества кластеров")
plt.axvline(optimal_k, color='r', linestyle='--', label=f'Оптимальное k: {optimal_k}')
plt.legend()
plt.show()
```

```python
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
kmeans.fit(X_pca)
```

```python
centroids = kmeans.cluster_centers_
print("Координаты центроидов кластеров:")
print(centroids)
```

```python
labels, counts = np.unique(kmeans.labels_, return_counts=True)
cluster_counts = dict(zip(labels, counts))
print("Количество вхождений в кластеры:")
print(cluster_counts)
```

```python
# Добавление меток кластеров в исходные данные
df["Cluster"] = kmeans.labels_
```

```python
df
```

```python
cluster_means = df.groupby("Cluster").mean(numeric_only=True)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_pca)

df_pca = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
df_pca["Cluster"] = kmeans.labels_

plt.figure(figsize=(8, 6))
sns.scatterplot(x="PC1", y="PC2", hue=df_pca["Cluster"], palette="viridis", data=df_pca, alpha=0.7)
plt.scatter(pca.transform(centroids)[:, 0], pca.transform(centroids)[:, 1], color='red', marker='X', s=200, label='Центроиды')    
plt.title("Визуализация кластеров (PCA)")
plt.xlabel("Главная компонента 1")
plt.ylabel("Главная компонента 2")
plt.legend()
plt.show()
```
